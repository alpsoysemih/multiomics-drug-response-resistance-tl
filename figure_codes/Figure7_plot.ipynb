```r
# read GDSC expression matrix and sample files

gdsc_experiment_matrix_gemcitabine <- read.delim(paste(pathfindR_dir, "GDSC_Experiment_Matrix_DNA_REPi_Gemcitabine_First_Strategy.tsv", sep = "/"))
colnames(gdsc_experiment_matrix_gemcitabine)[2:ncol(gdsc_experiment_matrix_gemcitabine)] <- sub("^X", "", colnames(gdsc_experiment_matrix_gemcitabine)[2:ncol(gdsc_experiment_matrix_gemcitabine)])
rownames(gdsc_experiment_matrix_gemcitabine) <- gdsc_experiment_matrix_gemcitabine$symbol
gdsc_experiment_matrix_gemcitabine <- gdsc_experiment_matrix_gemcitabine[,-1]
gdsc_experiment_matrix_gemcitabine <- as.matrix(gdsc_experiment_matrix_gemcitabine)
colnames(gdsc_experiment_matrix_gemcitabine) <- as.integer(colnames(gdsc_experiment_matrix_gemcitabine))

gdsc_cases_data_gemcitabine <- read.delim(paste(pathfindR_dir, "GDSC_DNA_REPi_Gemcitabine_Samples_First_Strategy.tsv", sep ="/"))

filter <- (gdsc_cases_data_gemcitabine["Response"] == 0)
cases_gemcitabine <- as.character(gdsc_cases_data_gemcitabine[filter, "Sample"])

score_matrix_gemcitabine <- score_terms(
  enrichment_table = save_filtered_gemcitabine,
  exp_mat = gdsc_experiment_matrix_gemcitabine,
  cases = cases_gemcitabine,
  use_description = TRUE, # default FALSE
  label_samples = FALSE, # default = TRUE
  case_title = "Resistant", # default = "Case"
  control_title = "Sensitive", # default = "Control"
  low = "#f7797d", # default = "green"
  mid = "#fffde4", # default = "black"
  high = "#1f4037" # default = "red"
)
```

```r
# read GDSC expression matrix and sample files

gdsc_experiment_matrix_fluorouracil <- read.delim(paste(pathfindR_dir, "GDSC_Experiment_Matrix_DNA_REPi_Fluorouracil_First_Strategy.tsv", sep = "/"))
colnames(gdsc_experiment_matrix_fluorouracil)[2:ncol(gdsc_experiment_matrix_fluorouracil)] <- sub("^X", "", colnames(gdsc_experiment_matrix_fluorouracil)[2:ncol(gdsc_experiment_matrix_fluorouracil)])
rownames(gdsc_experiment_matrix_fluorouracil) <- gdsc_experiment_matrix_fluorouracil$symbol
gdsc_experiment_matrix_fluorouracil <- gdsc_experiment_matrix_fluorouracil[,-1]
gdsc_experiment_matrix_fluorouracil <- as.matrix(gdsc_experiment_matrix_fluorouracil)
colnames(gdsc_experiment_matrix_fluorouracil) <- as.integer(colnames(gdsc_experiment_matrix_fluorouracil))

gdsc_cases_data_fluorouracil <- read.delim(paste(pathfindR_dir, "GDSC_DNA_REPi_Fluorouracil_Samples_First_Strategy.tsv", sep ="/"))

filter <- (gdsc_cases_data_fluorouracil["Response"] == 0)
cases_fluorouracil <- as.character(gdsc_cases_data_fluorouracil[filter, "Sample"])

score_matrix_fluorouracil <- score_terms(
  enrichment_table = save_filtered_fluorouracil,
  exp_mat = gdsc_experiment_matrix_fluorouracil,
  cases = cases_fluorouracil,
  use_description = TRUE, # default FALSE
  label_samples = FALSE, # default = TRUE
  case_title = "Resistant", # default = "Case"
  control_title = "Sensitive", # default = "Control"
  low = "#f7797d", # default = "green"
  mid = "#fffde4", # default = "black"
  high = "#1f4037" # default = "red"
)
```
