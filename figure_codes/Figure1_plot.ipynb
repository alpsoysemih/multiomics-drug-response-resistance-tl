```python
# List of images
images = [
    mitosis_image1, mitosis_image2, mitosis_image3, mitosis_image4, mitosis_image5,
    cyto_image1, cyto_image2, fluorouracil_image3, fluorouracil_image4, fluorouracil_image5,
    fluorouracil_image1, fluorouracil_image2, gemcitabine_image3, gemcitabine_image4, gemcitabine_image5,
    cetuximab_image1, cetuximab_image2, cetuximab_image3, cetuximab_image4, cetuximab_image5
]

# Define grid size
rows, cols = 4, 5

# Create the figure with larger size for better visibility
fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4), gridspec_kw={'hspace': 0.02, 'wspace': 0.02})
axes = axes.flatten()  # Flatten the grid for easier indexing

# Remove space between plots entirely
plt.subplots_adjust(wspace=0, hspace=0, left=0.02, right=0.98, top=0.9, bottom=0.02)

# Add each image to a subplot
for ax, img in zip(axes, images):
    ax.imshow(img)  # Display the image
    ax.axis('off')  # Turn off axes

# Hide any unused axes if there are extra slots
for ax in axes[len(images):]:
    ax.axis('off')

# Add column tags
column_labels = ['A.', 'B.', 'C.', 'D.', 'E.']  # Adjust for the number of columns
for i, label in enumerate(column_labels):
    # Dynamically calculate the x-coordinate for each column tag
    x_position = 0.04 + i * (0.92 / cols)  # Adjust spacing if needed
    fig.text(x_position, 0.92, label, ha='center', va='center', fontsize=17, fontweight='bold')

# Save the figure
plt.savefig(save_fig, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
