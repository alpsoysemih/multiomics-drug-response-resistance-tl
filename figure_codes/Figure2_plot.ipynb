```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
import warnings

# suppress all warnings
warnings.filterwarnings("ignore")
```

```python
# define directories

mitosis_path = "/Volumes/Expansion/Thesis Work/Results/GDSC_DEGs_inhibitors/MITOSISi/"
cytoskeleton_path = "/Volumes/Expansion/Thesis Work/Results/GDSC_DEGs_inhibitors/CYTOi/"
dna_replication_path = "/Volumes/Expansion/Thesis Work/Results/GDSC_DEGs_inhibitors/DNA_REPi/"
egfr_path = "/Volumes/Expansion/Thesis Work/Results/GDSC_DEGs_inhibitors/EGFRi/"
output_path = "/Users/semihalpsoy/Desktop/Paper_Drug_Resistance_Mechanisms/Figures/Figure 2.jpeg"
```

### 1. Mitotic inhibitors

```python
# Load the data

data = pd.ExcelFile(mitosis_path + "MITOSISi_Differentially_Expressed_Genes (EnsemblID).xlsx")
sheet_names = [s for s in data.sheet_names if s != "Common DEGs"]

results = []
for sheet in sheet_names:
    df = pd.read_excel(data, sheet_name=sheet)
    if "logFC" in df.columns and "adj.P.Val" in df.columns:
        filtered = df[(df["adj.P.Val"] < 0.05) & (df["logFC"].abs() > 1)]
        results.append([sheet, filtered.shape[0]])

data = pd.DataFrame(results, columns=["Drugs", "Number"])

# Set colors: red for Paclitaxel, skyblue for others
colors = ['red' if drug.lower() == 'paclitaxel' else 'skyblue' for drug in data['Drugs']]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bars
bars = ax.bar(data['Drugs'], data['Number'], color=colors)

# Annotate each bar with the corresponding number of DEGs
for bar in bars:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # Center the text
        yval + 1,  # Position slightly above the bar
        f'{int(yval)}',  # The number of DEGs
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

# Add text annotation on the plot
ax.text(
    1.5,  # x-coordinate (relative to plot)
    max(data['Number']) - 50,  # y-coordinate above the highest bar
    'Number of unionized DEGs\nwith mitotic inhibitors: 996',
    ha='center', va='center', fontsize=16, fontweight='bold'
)

# Customize the plot
#ax.set_xlabel('Drugs', fontsize=16, fontweight='bold')  # Bold x-label
ax.set_ylabel('Number of DEGs', fontsize=17, fontweight='bold')  # Bold y-label
ax.set_title('Number of DEGs for mitotic inhibitors', fontsize=17, fontweight='bold')  # Bold title
ax.set_xticklabels(data['Drugs'], rotation=45, ha='right')
fig.tight_layout()

# Assign the figure to a variable
mitosis_figure = fig

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define annotation texts for each subplot
annotation_texts = {
    "Mitotic inhibitors": "Number of unionized DEGs\nwith mitotic inhibitors: 996",
    "Cytoskeleton inhibitors": "Number of unionized DEGs\nwith cytoskeleton inhibitors: 845",
    "DNA replication inhibitors": "Number of unionized DEGs\nwith DNA replication inhibitors: 858",
    "EGFR signaling inhibitors": "Number of unionized DEGs\nwith EGFR signaling inhibitors: 1028"
}

# Titles for each subplot (must correspond to stored figures)
titles = ["Mitotic inhibitors", "Cytoskeleton inhibitors", "DNA replication inhibitors", "EGFR signaling inhibitors"]

# List of stored figure objects
figures = [mitosis_figure, cytoskeleton_figure, dna_replication_figure, egfr_figure]

# Subplot tags
subplot_tags = ["A.", "B.", "C.", "D."]

# Create a 2x2 grid for subplots
fig, axes = plt.subplots(2, 2, figsize=(32, 16))
axes = axes.flatten()

# Loop through the stored figures and draw them into the new grid
for i, (figure, ax) in enumerate(zip(figures, axes)):
    # Extract the first axis from each stored figure
    original_ax = figure.axes[0]
    
    # Copy bars from the original figure
    for bar in original_ax.patches:
        ax.add_patch(plt.Rectangle(
            (bar.get_x(), bar.get_y()), bar.get_width(), bar.get_height(),
            facecolor=bar.get_facecolor(), edgecolor=bar.get_edgecolor()
        ))

        # Copy bar height and add integer annotation
        yval = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            yval + 5,  # Slightly above the bar
            f'{int(yval)}',  # Convert to integer
            ha='center', va='bottom', fontsize=18, fontweight='bold', color='black'
        )
    
    # Set title, x-label, and (if first column) y-label
    ax.set_title(titles[i], fontsize=22, fontweight="bold")
    ax.set_xlabel(original_ax.get_xlabel(), fontsize=18, fontweight="bold")
    if i % 2 == 0:
        ax.set_ylabel(original_ax.get_ylabel(), fontsize=18, fontweight="bold")
    
    # Copy x and y limits
    ax.set_xlim(original_ax.get_xlim())
    ax.set_ylim(original_ax.get_ylim())
    
    # Copy tick labels and rotations
    ax.set_xticks(original_ax.get_xticks())
    ax.set_xticklabels(original_ax.get_xticklabels(), rotation=45, ha='right', fontsize=16)
    ax.set_yticks(original_ax.get_yticks())
    ax.set_yticklabels(original_ax.get_yticklabels(), fontsize=16)
    
    # Copy legend if present
    if original_ax.get_legend():
        ax.legend(
            loc='best', title=original_ax.get_legend().get_title().get_text(),
            fontsize=16, title_fontsize=18
        )
    
    # Adjust annotation position: Left for the third plot, Right for others
    ann_text = annotation_texts[titles[i]]
    ann_x_position = 0.3 if i == 1 else 0.25 if i == 2 else 0.4 if i == 3 else 0.8    
    ax.text(ann_x_position, 0.95, ann_text,
            transform=ax.transAxes,
            ha='center', va='top', fontsize=18, fontweight='bold')

    # Add subplot tags (a., b., c., d.) in the top-left corner
    ax.text(-0.05, 1.1, subplot_tags[i],
            transform=ax.transAxes,
            fontsize=27, fontweight="bold", va="top", ha="right")

# Adjust layout spacing
plt.tight_layout()
fig.subplots_adjust(wspace=0.1, hspace=0.4)

# Save the combined figure
fig.savefig(output_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
